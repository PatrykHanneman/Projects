```python
# Importowanie niezbędnych bibliotek
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# usunięcie ewentualnych futurewarningów
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# Wczytanie wyników z pliku CSV
df = pd.read_csv("wyniki.csv", header=None, names=["name", "attempts"])

# Obliczenie średniej liczby prób
average_attempts = df['attempts'].mean()

# Wizualizacja danych za pomocą seaborn
plt.figure(figsize=(10, 6))

# Tworzymy wykres słupkowy z seaborn
sns.barplot(x='name', y='attempts', data=df, palette="viridis")

# Dodanie linii średniej liczby prób
plt.axhline(average_attempts, color='r', linestyle='--', label=f'Average attempts: {average_attempts:.2f}')

# Ustawienie tytułu i etykiet osi
plt.title('Number of attempts per player', fontsize=16)
plt.xlabel('Player', fontsize=12)
plt.ylabel('Number of attempts', fontsize=12)
plt.xticks(rotation=45)

# Dodanie legendy
plt.legend()

# Ustawienie layoutu, aby elementy nie nakładały się
plt.tight_layout()

# Wyświetlanie wykresu
plt.show()

# Wyświetlanie średniej liczby prób w konsoli
print(f"Average attempts: {average_attempts:.2f}")
```
