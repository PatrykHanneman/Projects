```python
# Import necessary libraries
from collections import Counter

import numpy as np
```

```python
with open('data.txt', 'r') as file:
    data = file.read()

# Parse the numbers into a list 
numbers = list(map(float, data.split()))  # Convert strings to floats

# Display the parsed numbers (optional, for verification)
numbers
```

# Count Number Frequencies
Use the Counter class from the collections module to count the frequency of each number.

```python
# Count the frequency of each number using Counter
number_counts = Counter(numbers)

# Find the most common numbers
most_common_numbers = number_counts.most_common()
```

```python
numbers.sort()
```

# Mean

```python
# with use of numpy
mean = np.mean(numbers)
print(mean)
```

```python
# without numpy
mean = sum(numbers) / len(numbers)
print(mean)
```

# Display Most Frequent Number
Sort the numbers by frequency and display the most frequently occurring one.

```python
# Extract the number with the highest frequency
max_frequency = most_common_numbers[0][1]  # Get the highest frequency
most_frequent_number = [num for num, freq in most_common_numbers if freq == max_frequency]

# Display the most frequently occurring number
most_frequent_number
```
